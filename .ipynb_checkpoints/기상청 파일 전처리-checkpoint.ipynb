```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('weather/sun/단구동_기온_202101_202112.csv', encoding='cp949')
data
```

```python
data.columns=["dayy", "hour", "value"]
data = data.dropna()
data.info()
```

```python
day = [31,28,31,30,31,30,31,31,30,31,30,31]
count = 0
data["month"] = 0

for i in data.index:
    print(data["dayy"][i])
    if data["dayy"][i] == " 1" and data["hour"][i] == 0.0:
        count += 1
    if count == 13:
        break
    data["month"][i] = count
data
```

```python
# data["month"]가 0이 아닌 행 추출
# mask = data["month"].isin([0])
# data[~mask] 
# data = data[~mask]


data.info()
```

```python
data = data[['month','dayy','value','hour']]
data
```

```python
data = data.groupby(['month','dayy']).mean()
data = data.sort_values(by=['month'])
data
```

```python
data.to_csv("weather/sun/단구동기온.csv",mode="w")
```
