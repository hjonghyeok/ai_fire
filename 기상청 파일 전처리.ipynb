```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('파일명', encoding='cp949')
data
```

```python
data.columns=["dayy", "hour", "value"]
data = data.dropna()
data.info()
```

```python
count = 0
data["month"] = 0

for i in data.index:
    if i == 0:
        aa = data["dayy"][i]
        bb = data["hour"][i]
    if data["dayy"][i] == aa and data["hour"][i] == bb:
        count += 1
    if count == 13:
        break
    data["month"][i] = count
data
```

```python
# data["month"]가 0이 아닌 행 추출
# mask = data["month"].isin([0])
# data[~mask] 
# data = data[~mask]
data = data[['month','dayy','value','hour']]
data = data.groupby(['month','dayy']).sum()
data = data.sort_values(by=['month'])
data
```

```python
data.to_csv("파일명",mode="w")
```
